```python
import pandas as pd 
train_data_original = pd.read_csv("train_data.csv")
train_data_updated = pd.read_csv("train_data_22_features.csv")
train_data_updated = train_data_updated.drop(['Unnamed: 0.1', 'Unnamed: 0'], axis=1)
train_data_updated
```

```python
data_test_1 = pd.read_csv('test_data.csv')
data = train_data_updated
data
```

```python
data.columns = data.columns.str.strip()
```

```python
data_2 = data.drop(['cluster','Bankrupt?'],axis=1)
```

```python
data_test_1.columns = data_test_1.columns.str.strip()

data_test_2 = data_test_1[data_2.columns]
```

```python
data_test_2
```

```python
# Before transformation
from sklearn.preprocessing import QuantileTransformer
import matplotlib.pyplot as plt
from scipy.stats import normaltest
from scipy.stats import probplot


for feature in data_test_2.columns:

    X = data_test_2[[feature]]
    # Instantiate the QuantileTransformer
    quantile_transformer = QuantileTransformer(output_distribution='normal', n_quantiles=500, random_state=0)

    # Reshape data for transformation (required for single features)
    data_reshaped = X.values.reshape(-1, 1)

    # Apply the Quantile Transformation
    X_normalized_data = quantile_transformer.fit_transform(data_reshaped).flatten()

    data_test_2[feature] = X_normalized_data
    
data_test_2
```

```python
data
```

```python
from sklearn.model_selection import train_test_split, KFold, cross_val_score
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import pandas as pd
import numpy as np
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import GridSearchCV, StratifiedKFold
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
from imblearn.over_sampling import SMOTE
from sklearn.preprocessing import StandardScaler



# Define features and target
X = data.drop(['Bankrupt?', 'cluster'], axis=1)
y_train = data['Bankrupt?']

# Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

smote = SMOTE(random_state=42)
X_train_resampled, y_train_new = smote.fit_resample(X_scaled, y_train)

X_train =X_train_resampled
X_test = data_test_2

# Stratified k-fold cross-validation setup
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

# Define the GaussianNB model
gaussian_nb = GaussianNB()

# Parameters grid for hyperparameter tuning
param_grid = {
    'var_smoothing': np.logspace(0,-9, num=100)
}

# GridSearchCV for hyperparameter tuning within the cross-validation
grid_search = GridSearchCV(
    estimator=gaussian_nb,
    param_grid=param_grid,
    scoring='recall',  # Can use 'accuracy' or other metrics
    n_jobs=-1,
    cv=skf,
    verbose=3
)

# Fit GridSearchCV on the resampled training data
grid_search.fit(X_train, y_train_new)

# Extract the best model
best_model = grid_search.best_estimator_

# Evaluate the best model on the untouched test set
predictions = best_model.predict(X_test)


# Perform K-fold cross-validation
cv_scores = cross_val_score(gaussian_nb, X_train, y_train_new, cv=skf, scoring='accuracy')
cv_recall_scores = cross_val_score(gaussian_nb, X_train, y_train_new, cv=skf, scoring='recall')

print("Cross-validation Accuracy scores:", cv_scores)
print("Cross-validation Recall scores:", cv_recall_scores)

# Fit model on entire training data
gaussian_nb.fit(X_train, y_train_new)
# Predict on test data
final_predictions = gaussian_nb.predict(X_test)


unique_values, counts = np.unique(final_predictions, return_counts=True)

print("Unique values:", unique_values)
print("Counts:", counts)


```
